```python
from fastmcp import FastMCP
import threading, asyncio

# Initialize the server with a name
mcp = FastMCP("WeatherService")

@mcp.tool()
def get_weather(city: str) -> str:
    """
    Get the current weather for a specific city.
    """
    # In a real app, you would call a weather API here.
    mock_data = {
        "new york": "Sunny, 72В°F",
        "london": "Rainy, 15В°C",
        "tokyo": "Cloudy, 22В°C"
    }
    return mock_data.get(city.lower(), f"Weather data for {city} is unavailable.")

# Run the server using the standard input/output transport
if __name__ == "__main__":
    threading.Thread(target=lambda: asyncio.run(mcp.run(transport="stdio")), daemon=True).start()
```
