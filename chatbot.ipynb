```python
# Basic Chat prompt to Ollama LLM
import ollama

response = ollama.chat(model="llama3.2", messages=[{"role": "user", "content": "What is the capital of France?"}])
print(response.message.content, response.model,"\n",response)
```

```python
import ollama

response = ollama.chat(model='llama3.2', messages=[
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
])
print(response['message']['content'])

```

```python
## Stream response example

stream = ollama.chat(
    model='llama3.2',
    messages=[{'role': 'user', 'content': 'Tell me a story.'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```

```python
import ollama

messages = [
    {
        "role": "user",
        "content": "Bot: Hello! How can I assist you today?\n"
    }
]




response = ollama.chat(model="llama3.2", messages=messages)
print(response['message']['content'])


while True:
    user_input = input("You: ")
    if not user_input:
        break  # exit loop on empty input
    messages.append({"role": "user", "content": user_input})
    response = ollama.chat(model="llama3.2", messages=messages)
    answer = response.message.content
    print("Bot:", answer)
    messages.append({"role": "assistant", "content": answer})
```
